```python
import warnings
import sys
import numpy as np
import pandas as pd
from IPython.display import display
warnings.simplefilter(action='ignore', category=FutureWarning)
excel_file_path ='/content/sample_data/FormattedData (4).xlsx'
xls = pd.ExcelFile(excel_file_path)
def calc_change(row):
    if row["Views_curr"] == row["Max Views"]:
        if row["Prev Max Views"] == 0:
            return 0
        else:
            return (row["Views_curr"] - row["Prev Max Views"]) / row["Prev Max Views"]
    else:
        if row["Max Views"] == 0:
            return 0
        else:
            return (row["Views_curr"] - row["Max Views"]) / row["Max Views"]
formattedJournalistData = pd.DataFrame(columns=["Name", "No of Shows", "Views", "Channel", "Date", "Designation", "Max Views", "Change of Views"])
df = pd.read_excel(xls, sheet_name="Journalist")
df = df.dropna(how="all").reset_index(drop=True)
az = df
az.columns  = az.columns.to_list()
channels = ["Azaad Digital", "Pukhtun Digital", "Kashmir Digital", "Burak"]
designations = ["Analyst", "Anchor", "Reporter"]
print("Starting: journalist data")
for channel in channels:
    print(f"now i will run {channel}")
    for designation in designations:
        print(f"now i will run {designation}")
        for i in range(1,12):
            prevDate = f"2025-09-{str(i-1).zfill(2)}"
            currDate = f"2025-09-{str(i).zfill(2)}"
            nextDate = f"2025-09-{str(i+1).zfill(2)}"
            dataPreviousDate = az[(az["Channel"] == channel) & (az["Designation"] == designation) & (az["Date"] == prevDate)]
            dataCurrentDate = az[(az["Channel"] == channel) & (az["Designation"] == designation) & (az["Date"] == currDate)]
            dataNextDate = az[(az["Channel"] == channel) & (az["Designation"] == designation) & (az["Date"] == nextDate)]
            if dataCurrentDate.empty:
                print(f"No data for {channel}, {designation}, {currDate}")
                break
            else:
                if currDate == "2025-09-01":
                    newData = pd.DataFrame({
                        "Name": dataCurrentDate["Name"],
                        "No of Shows": dataCurrentDate["No of Shows"],
                        "Views": dataCurrentDate["Views"],
                        "Channel": dataCurrentDate["Channel"],
                        "Date": dataCurrentDate["Date"],
                        "Designation": dataCurrentDate["Designation"],
                        "Max Views": dataCurrentDate["Views"],
                        "Change of Views": 0
                    })
                    formattedJournalistData = pd.concat([formattedJournalistData, newData], ignore_index=True)
                else:
                    merged = pd.merge(
                        dataCurrentDate[["Name", "Views", "No of Shows", "Channel", "Date", "Designation"]],
                        dataPreviousDate[["Name", "Views", "No of Shows", "Channel", "Date", "Designation"]],
                        on="Name",
                        how="inner",
                        suffixes=("_curr", "_prev")
                    )
                    merged["Max Views"] = merged[["Views_curr", "Views_prev"]].max(axis=1)
                    pmv = formattedJournalistData[(formattedJournalistData["Channel"] == channel) & (formattedJournalistData["Designation"] == designation) & (formattedJournalistData["Date"] == prevDate)]
                    merged = merged.merge(
                        pmv[["Name", "Max Views"]],
                        on="Name",
                        how="left",
                        suffixes=("", "_prevMax")
                    )
                    merged.rename(columns={"Max Views_prevMax": "Prev Max Views"}, inplace=True)
                    merged["Change of Views"] = merged.apply(calc_change, axis=1)
                    newData = pd.DataFrame({
                        "Name": merged["Name"],
                        "No of Shows": merged["No of Shows_curr"],
                        "Views": merged["Views_curr"],
                        "Channel": merged["Channel_curr"],
                        "Date": merged["Date_curr"],
                        "Designation": merged["Designation_curr"],
                        "Max Views": merged["Max Views"],
                        "Change of Views": merged["Change of Views"]
                    })
                    formattedJournalistData = pd.concat([formattedJournalistData, newData], ignore_index=True)



print("completed journalist data")



```

```python
print("\n\nstarting vlogger data")
formattedVloggerData = pd.DataFrame(columns=["Name", "No of Vlogs", "Views", "Platform", "Channel", "Date", "Max Views", "Change of Views"])
df = pd.read_excel(xls, sheet_name="Vlogger")
df = df.dropna(how="all").reset_index(drop=True)
az = df
az.columns  = az.columns.to_list()
channels = ["Azaad Digital", "Pukhtun Digital", "Kashmir Digital", "Burak"]
platforms = ["Youtube", "Twitter (X)", "TikTok", "Facebook", "Instagram"]
for channel in channels:
    for platform in platforms:
        print(f"now i will run {channel} for {platform}")
        for i in range(1,12):
            prevDate = f"2025-09-{str(i-1).zfill(2)}"
            currDate = f"2025-09-{str(i).zfill(2)}"
            nextDate = f"2025-09-{str(i+1).zfill(2)}"
            dataPreviousDate = az[(az["Channel"] == channel) & (az["Platform"] == platform) & (az["Date"] == prevDate)]
            dataCurrentDate = az[(az["Channel"] == channel) & (az["Platform"] == platform) & (az["Date"] == currDate)]
            dataNextDate = az[(az["Channel"] == channel) & (az["Platform"] == platform) & (az["Date"] == nextDate)]
            if dataCurrentDate.empty:
                print(f"No data for {channel}, {platform}, {currDate}")
                continue
            else:
                if currDate == "2025-09-01":
                    newData = pd.DataFrame({
                        "Name": dataCurrentDate["Name"],
                        "No of Vlogs": dataCurrentDate["No of Vlogs"],
                        "Views": dataCurrentDate["Views"],
                        "Channel": dataCurrentDate["Channel"],
                        "Date": dataCurrentDate["Date"],
                        "Platform": dataCurrentDate["Platform"],
                        "Max Views": dataCurrentDate["Views"],
                        "Change of Views": 0
                    })
                    formattedVloggerData = pd.concat([formattedVloggerData, newData], ignore_index=True)
                else:
                    merged = pd.merge(
                        dataCurrentDate[["Name", "Views", "No of Vlogs", "Channel", "Date", "Platform"]],
                        dataPreviousDate[["Name", "Views", "No of Vlogs", "Channel", "Date", "Platform"]],
                        on="Name",
                        how="inner",
                        suffixes=("_curr", "_prev")
                    )
                    merged["Max Views"] = merged[["Views_curr", "Views_prev"]].max(axis=1)
                    pmv = formattedVloggerData[(formattedVloggerData["Channel"] == channel) & (formattedVloggerData["Platform"] == platform) & (formattedVloggerData["Date"] == prevDate)]
                    merged = merged.merge(
                        pmv[["Name", "Max Views"]],
                        on="Name",
                        how="left",
                        suffixes=("", "_prevMax")
                    )
                    merged.rename(columns={"Max Views_prevMax": "Prev Max Views"}, inplace=True)
                    merged["Change of Views"] = merged.apply(calc_change, axis=1)
                    newData = pd.DataFrame({
                        "Name": merged["Name"],
                        "No of Vlogs": merged["No of Vlogs_curr"],
                        "Views": merged["Views_curr"],
                        "Channel": merged["Channel_curr"],
                        "Date": merged["Date_curr"],
                        "Platform": merged["Platform_curr"],
                        "Max Views": merged["Max Views"],
                        "Change of Views": merged["Change of Views"]
                    })
                    formattedVloggerData = pd.concat([formattedVloggerData, newData], ignore_index=True)
from google.colab import files
from openpyxl.styles import Alignment, Border, Side
from openpyxl.utils import get_column_letter

with pd.ExcelWriter("CompletelyFormattedData.xlsx", engine="openpyxl") as writer:
    formattedVloggerData.to_excel(writer, sheet_name="Vlogger", index=False)
    formattedJournalistData.to_excel(writer, sheet_name="Journalist", index=False)

    workbook = writer.book
    thin_border = Border(
        left=Side(style="thin"),
        right=Side(style="thin"),
        top=Side(style="thin"),
        bottom=Side(style="thin")
    )
    for sheetname in writer.sheets:
        worksheet = writer.sheets[sheetname]

        max_row = worksheet.max_row
        max_col = worksheet.max_column

        for row in worksheet.iter_rows(min_row=1, max_row=max_row, min_col=1, max_col=max_col):
            for cell in row:
                cell.alignment = Alignment(horizontal="center", vertical="center")
                cell.border = thin_border

        for col in range(1, max_col + 1):
            col_letter = get_column_letter(col)
            worksheet.column_dimensions[col_letter].width = 25

        for row in range(1, max_row + 1):
            worksheet.row_dimensions[row].height = 25

files.download("CompletelyFormattedData.xlsx")
```
